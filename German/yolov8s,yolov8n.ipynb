```python
from google.colab import files
from PIL import Image
import io
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Step 1: Upload image
uploaded = files.upload()

# Step 2: Load the model (make sure this path is to your best .pt file)
model = YOLO("/content/drive/MyDrive/German/lightweight_models/yolov8s_augmented_v22/weights/best.pt")

# Step 3: Read the uploaded image into memory
for name, file in uploaded.items():
    img = Image.open(io.BytesIO(file)).convert("RGB")

    # Step 4: Run prediction
    results = model.predict(img)

    # Step 5: Show result
    res_img = results[0].plot()  # Get rendered result
    plt.imshow(res_img)
    plt.axis('off')
    plt.title("Prediction")
    plt.show()
```
